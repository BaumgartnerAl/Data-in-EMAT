<h4>Bar Graph</h4>

```python
df_likes.groupby('title').count().sort_values('media_list_data', ascending=False).head(20).plot.bar(y='media_list_data', fontsize=(12), ylabel='Liked Posts', xlabel='Account Name', figsize = (10, 5))
```

Above is a bar graph showing the number of liked posts by the top 20 accounts. Visualizing the data in this way is interesting to look at. After the top 20 or so posts, each account has only 1 or 2 liked posts which I think could be valuable information to analyze. 

From the top 20 accounts, 17 of them are art accounts. Of those 17, 7 of those accounts I recognize based on name. The 3 non-art accounts I also recognize from name. From this, we can see that about 50% of these top 20 accounts have left a positive impression on me, as I am able to recognize the account and picture in my mind some of their work. This indicates that, in most cases, liked posts are a good indication of accounts that have left a positive impact on any given user.
